# NumPy — Array Properties and Operations

This notebook covers:
- Array structure and metadata  
- Data types and casting  
- Arithmetic operations  
- Aggregation functions  


```python
import numpy as np
```

## Array Properties


```python
arr = [
    [1, 4, 6, 8],
    [2, 4, 7, 9],
    [1, 2, 3, 4]
]
arr = np.array(arr)
arr
```

### shape — Number of rows and columns


```python
arr.shape
```

### size — Total number of elements


```python
arr.size
```

### ndim — Number of dimensions
A single scalar has 0 dimensions,  
a 1D array has 1 dimension,  
a 2D matrix has 2 dimensions, and so on.


```python
arr.ndim
```

## Data Types and Type Casting
NumPy commonly uses:
- int32, int64  
- float32, float64  
- bool  
- complex128  

NumPy automatically infers a data type unless specified.


### dtype — Check array data type


```python
arr.dtype
```

### astype — Convert array data type  
Note: casting is applied to the entire array.


```python
arr = arr.astype(str)
arr.dtype
# Convert back to integers for further operations.
arr = arr.astype(int)
arr
```

## Arithmetic Operations  
Operations are applied element-wise.


```python
arr + 5
arr - 5
arr * 5
arr / 5
arr ** 5
```

## Aggregation Functions


```python
np.sum(arr)
np.mean(arr)
np.min(arr)
np.max(arr)
np.std(arr)
np.var(arr)
```
