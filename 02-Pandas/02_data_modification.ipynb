This module covers:
- Adding new columns  
- Inserting columns at a specific position  
- Updating values

```python
import pandas as pd

df = pd.read_csv("../Datasets/employees.csv", encoding="latin1")
dataset = df.head(10)
dataset2 = df.tail(10)

dataset
```

## Add a new column at the end

```python
dataset["AGE"] = 26
dataset
```

## Insert "City" at index 0

```python
dataset.insert(0, "City", "Hyderabad")
dataset
```

## Update a specific value using `.loc`

```python
dataset.loc[0, "City"] = "Constantinpole"
dataset
```

## Delete the "First Name" column

```python
dataset.drop(columns=["First Name"], inplace=True)
dataset
```
